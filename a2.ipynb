```python
import pandas as pd
import plotly.express as px

# Data for Isha M Ambani's shareholding


# Convert to DataFrame
df = pd.DataFrame(dee)

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Plotting with more detailed elements
fig = px.line(df, x="Date", y="Share", color='BSE Code',
              markers=True,  # Add markers to each data point
              hover_data={"Date": "|%B %d, %Y", "BSE Code": True, "Share": ":.2f"},
              title='Isha M Ambani Shareholding Trends Over Time')


fig.update_xaxes(
    dtick="M3",  # Every 3 months
    tickformat="%b\n%Y",  # Show abbreviated month and year
    ticklabelmode="period"
)


fig.update_traces(mode="lines+markers", line_shape='linear')  # Display lines with markers


fig.update_layout(
    title={
        'text': "Isha M Ambani's Shareholding Trends (2021-2024)",
        'y': 0.9,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title="Date",
    yaxis_title="Shareholding (%)",
    legend_title="BSE Code",
    hovermode='x unified'  # Show hover information unified across points
)

# Show the plot
fig.show()
```
